## Installation of nnunet as a package
+ Following instructions from the [nnunet github repository](https://github.com/MIC-DKFZ/nnUNet#installation). It appeared that if pytorch needs to be compiled better use conda.

+ Hence I reproduced steps described in (https://towardsdatascience.com/conda-google-colab-75f7c867a522) 

```
%%bash
# select latest miniconda installer with python 3 version
MINICONDA_INSTALLER_SCRIPT=Miniconda3-latest-Linux-x86_64.sh
# install path
MINICONDA_PREFIX=/usr/local
# download miniconda installer
wget https://repo.continuum.io/miniconda/$MINICONDA_INSTALLER_SCRIPT
# add execution rights so that installer can be launched
chmod +x $MINICONDA_INSTALLER_SCRIPT
# install miniconda
./$MINICONDA_INSTALLER_SCRIPT -b -f -p $MINICONDA_PREFIX
# add conda command to path
echo ". /usr/local/etc/profile.d/conda.sh" >> ~/.bashrc
# update conda command silently
conda update conda -y
```

```
%%bash
conda create -n test-nnunet python=3.6 pytorch
```
